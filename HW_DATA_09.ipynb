```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
```

```python

# Завантаження датасету Fashion MNIST
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python

def create_model():
    model = models.Sequential([
        layers.Flatten(input_shape=(28, 28)),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(10, activation='softmax')
    ])
    return model
```

```python
optimizers = ['adam', 'rmsprop', 'sgd', 'adadelta', 'adagrad']


histories = {}
for optimizer in optimizers:
    print(f"Навчання з оптимізатором {optimizer}...")
    model = create_model()
    history = train_model(model, optimizer=optimizer)
    histories[optimizer] = history
```

В такій формі краще не навчати модель, виконувалось більше 40хв і точність не відповідає нормі

```python
plt.figure(figsize=(10, 6))
for optimizer, history in histories.items():
    plt.plot(history.history['val_accuracy'], label=optimizer)
plt.title('Порівняння оптимізаторів')
plt.xlabel('Епохи')
plt.ylabel('Точність на валідаційному наборі')
plt.legend()
plt.show()
```

```python
# Функція для навчання моделі з оптимізатором Adam
def train_model_adam(model, epochs=100):
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2,
                              patience=5, min_lr=0.001)

    # early_stopping = EarlyStopping(monitor='val_accuracy', patience=6, restore_best_weights=True)

    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=256, validation_split=0.2, verbose=1, callbacks=[reduce_lr])

    return history


# Функція для навчання моделі з оптимізатором RMSprop
def train_model_rmsprop(model, epochs=100):
    model.compile(optimizer='rmsprop',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_accuracy', patience=6, restore_best_weights=True)

    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=64, validation_split=0.2, callbacks=[early_stopping])

    return history

# Функція для навчання моделі з оптимізатором SGD
def train_model_sgd(model, epochs=100):
    model.compile(optimizer='sgd',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_accuracy', patience=6, restore_best_weights=True)

    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=64, validation_split=0.2, callbacks=[early_stopping])

    return history

# Функція для навчання моделі з оптимізатором Adadelta
def train_model_adadelta(model, epochs=100):
    model.compile(optimizer='adadelta',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_accuracy', patience=6, restore_best_weights=True)

    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=64, validation_split=0.2, callbacks=[early_stopping])

    return history

# Функція для навчання моделі з оптимізатором Adagrad
def train_model_adagrad(model, epochs=100):
    model.compile(optimizer='adagrad',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_accuracy', patience=6, restore_best_weights=True)

    history = model.fit(train_images, train_labels, epochs=epochs, batch_size=64, validation_split=0.2, callbacks=[early_stopping])

    return history

def display_results(history, model, optimizer_name):
    test_loss, test_acc = model.evaluate(test_images, test_labels)
    print(f'Точність на тестових даних з оптимізатором {optimizer_name}:', test_acc)

    train_loss, train_acc = model.evaluate(train_images, train_labels)
    print(f'Точність на тренувальних даних з оптимізатором {optimizer_name}:', train_acc)

    plt.plot(history.history['accuracy'], label='Точність на тренувальному наборі')
    plt.plot(history.history['val_accuracy'], label='Точність на валідаційному наборі')
    plt.xlabel('Епохи')
    plt.ylabel('Точність')
    plt.title(f'Оптимізатор: {optimizer_name}')
    plt.legend()
    plt.show()

    plt.plot(history.history['loss'], label='Втрати на тренувальному наборі')
    plt.plot(history.history['val_loss'], label='Втрати на валідаційному наборі')
    plt.xlabel('Епохи')
    plt.ylabel('Втрати')
    plt.title(f'Оптимізатор: {optimizer_name}')
    plt.legend()
    plt.show()

# Навчання моделі з оптимізатором Adam
model_adam = create_model()
history_adam = train_model_adam(model_adam)
display_results(history_adam, model_adam, 'Adam')

# Навчання моделі з оптимізатором RMSprop
model_rmsprop = create_model()
history_rmsprop = train_model_rmsprop(model_rmsprop)
display_results(history_rmsprop, model_rmsprop, 'RMSprop')

# Навчання моделі з оптимізатором SGD
model_sgd = create_model()
history_sgd = train_model_sgd(model_sgd)
display_results(history_sgd, model_sgd, 'SGD')

# Навчання моделі з оптимізатором Adadelta
model_adadelta = create_model()
history_adadelta = train_model_adadelta(model_adadelta)
display_results(history_adadelta, model_adadelta, 'Adadelta')

# Навчання моделі з оптимізатором Adagrad
model_adagrad = create_model()
history_adagrad = train_model_adagrad(model_adagrad)
display_results(history_adagrad, model_adagrad, 'Adagrad')

```

```python
# Виведення графіка порівняння оптимізаторів
plt.figure(figsize=(10, 6))
plt.plot(history_adam.history['val_accuracy'], label='adam')
plt.plot(history_rmsprop.history['val_accuracy'], label='rmsprop')
plt.plot(history_sgd.history['val_accuracy'], label='sgd')
plt.plot(history_adadelta.history['val_accuracy'], label='adadelta')
plt.plot(history_adagrad.history['val_accuracy'], label='adagrad')
plt.title('Порівняння оптимізаторів')
plt.xlabel('Епохи')
plt.ylabel('Точність на валідаційному наборі (%)')
plt.legend()
plt.grid(True)
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()])

# Додавання значень точності під кожну лінію
for history in [history_adam, history_rmsprop, history_sgd, history_adadelta, history_adagrad]:
    plt.text(len(history.epoch)-1, history.history['val_accuracy'][-1], '{:.1%}'.format(history.history['val_accuracy'][-1]), fontsize=9, verticalalignment='bottom')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

class_names = [
    "T-shirt/top",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot"
]


random_indices = np.random.choice(test_images.shape[0], 5, replace=False)

# Display each image along with its predicted and true classes
for i, idx in enumerate(random_indices):
    image = test_images[idx]
    true_label = test_labels[idx]

    # Predict the class of the image
    predictions = model_adam.predict(np.expand_dims(image, axis=0))
    predicted_label = np.argmax(predictions)

    print(f"Image {i+1}:")
    print("True Class:", class_names[true_label])
    print("Predicted Class:", class_names[predicted_label])

    plt.figure(figsize=(2, 2))
    plt.imshow(image, cmap=plt.cm.binary)
    plt.xlabel(f"True Class: {class_names[true_label]}, Predicted Class: {class_names[predicted_label]}")
    plt.colorbar()
    plt.show()
```
